### 補足
- xの値が大きい場合は、exp(x)の値が大きくなりすぎて計算できなくなります
- そこで、最も大きなxを利用することで、安定した計算が可能になります。

```python
def softmax2(x):
    x_max = torch.max(x, dim=1, keepdim = True)[0]
    x = torch.exp(x - x_max)
    return x / torch.sum(x, dim=1, keepdim=True)
```
