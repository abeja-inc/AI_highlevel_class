```python
import torch

import torch.nn as nn
import torch.nn.functional as F

import numpy as np
```

### 課題
- 次のネットワークがどのような構成になっているかを考え、図示してみましょう

```python
class MLPNet(nn.Module):
    def __init__(self):
        super(MLPNet, self).__init__()
        self.fc1 = nn.Linear(3, 4)   
        self.fc2 = nn.Linear(4, 6)
        self.fc3 = nn.Linear(6, 8)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)
```
