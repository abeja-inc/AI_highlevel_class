### 課題1
- 以下の関数reluを完成させましょう
- pytorchにおいてはmaxはtorch.maxを使います。


```python
def relu(t):
    zeros = torch.zeros_like(t)
    #return torch.[A]([B], [C])
    return torch.max(zeros, t)
```

### 課題1の確認
- 上記relu関数が正しい結果を出力できているかを確認しましょう

```python
x = torch.rand(1, 3) - 0.5
print('x: ', x)
print()

y = relu(x)
print('relu(x): ', y)

len(x[0])
```

```python
x = np.array(range(-10, 10, 1))
x = torch.from_numpy(x).clone()
print('x: ', x)
print()

y = relu(x)
print('relu(x): ', y)
```

```python
# x座標の値
x0 = x

# y座標の値
x1 = y

# 簡易的なプロット法
plt.plot(x0, x1)

# グラフを表示
plt.show()
```

### 課題2
- 以下の関数softmaxを完成させましょう
- pytorchにおいてはexpはtorch.expを使います。


```python
def softmax(x):
#    x = torch.[A](x)
#    return [B] / torch.[C]([D], dim=1, keepdim=True)
    x = torch.exp(x)
    return x / torch.sum(x, dim=-1, keepdim=True)
```

### 課題2の確認
- 上記softmax関数が正しい結果を出力できているかを確認しましょう

```python
x = torch.rand(2, 3) - 0.5
print('x: ', x)

sx = softmax(x)
print('softmax(x):', sx)
```

### 補足
- xの値が大きい場合は、exp(x)の値が大きくなりすぎて計算できなくなります
- そこで、最も大きなxを利用することで、安定した計算が可能になります。

```python
def softmax2(x):
    x_max = torch.max(x, dim=-1, keepdim = True)[0]
    x = torch.exp(x - x_max)
    return x / torch.sum(x, dim=-1, keepdim=True)
```

```python
x = np.array(range(0, 50, 1))/10
#x = np.array([[x],[x]])
x = torch.from_numpy(x).clone()
print('x: ', x)

sx = softmax2(x)
print('softmax(x): ', sx)
```

```python
# x座標の値
x0 = x

# y座標の値
x1 = sx

# 簡易的なプロット法
plt.plot(x0, x1)

# グラフを表示
plt.show()
```
