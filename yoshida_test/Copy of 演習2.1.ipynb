```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
```

### 課題
- 以下のMomentumSGDの関数の[A]から[D]に当てはまる変数を入力しましょう

```python
# params・・・ネットワークのパラメータ
# d_p_list・・・パラメータの勾配
# momentum_buffer_list・・・前回のパラメータ更新量
# momentum・・・モーメンタムのパラメータ
# lr・・・・更新の係数
def sgd(params, d_p_list, momentum_buffer_list, momentum, lr):
    for i, param in enumerate(params):
        # i番目のパラメータの微分値をd_pとする
        d_p = d_p_list[i]

        # 前回の更新におけるi番目のパラメータの更新量をbufとする
        buf = momentum_buffer_list[i]

        # MomentumSGDの式より更新量を d_p + momemtum * bufで算出
        buf = momentum * buf + lr * d_p
        
        # パラメータの更新量をアップデート
        momentum_buffer_list[i] = buf
        
        # パラメータを更新（param = param - lr * buf）
        param.add_(buf, alpha=-buf[i])
```

### 課題の確認
- 上記sgd関数が正しい結果を出力できているかを確認しましょう

```python
params = torch.rand(2, 3)
print('params: ', params)

d_p_list = torch.rand(2, 3)
print('gradients: ', d_p_list)

momentum_buffer_list = torch.rand(2, 3)
print('momentum buffer: ', momentum_buffer_list)

sgd(params, d_p_list, momentum_buffer_list, 0.9, 0.1)
print('result: ', params)
```
